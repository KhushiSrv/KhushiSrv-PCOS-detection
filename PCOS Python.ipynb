```python
dataset = dataset[['PCOS (Y/N)', ' Age (yrs)', 'Weight (Kg)','Height(Cm) ','Cycle(R/I)', 'Cycle length(days)', 
                   'Hip(inch)','Waist(inch)','Weight gain(Y/N)','Skin darkening (Y/N)','Hair loss(Y/N)','Pimples(Y/N)','Reg.Exercise(Y/N)']]
```

```python
dataset
```

```python
attribute = dataset.drop('PCOS (Y/N)', axis = 1).values
target = dataset['PCOS (Y/N)']
```

```python
attribute
```

```python
target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
attribute = scaler.fit_transform(dataset)
attribute
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(attribute, target , test_size=0.20, shuffle=False)
```

```python
print(X_train.shape,
Y_train.shape,
X_test.shape,
Y_test.shape)
```

```python
import keras
```

```python
from keras.utils import to_categorical
Y_train = to_categorical(Y_train, num_classes=2)
Y_test = to_categorical (Y_test, num_classes=2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
model = Sequential()
model.add(Dense(13, activation='relu', input_shape=(13,)))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(10, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))
```

```python
optimizer = Adam(lr=0.001)  
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
history = model.fit(X_train, Y_train, batch_size=32, epochs=10, validation_split=0.1,callbacks=[early_stopping])
```

```python
model.summary()
```

```python
pred =  model.predict(X_test)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, Y_test)
print(f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}")
```

```python
import numpy as np 
from sklearn.metrics import confusion_matrix
predicted_labels = np.argmax(pred, axis=1)
true_labels = np.argmax(Y_test, axis=1)
cm = confusion_matrix(true_labels, predicted_labels)

print("Confusion Matrix:")
print(cm)
```

```python
tn, fp, fn, tp = cm.ravel() 
precision = 1.0 * tp / (tp + fp) 
recall = 1.0 * tp / (tp + fn) 
f1 = (2 * recall * precision) / (recall + precision) 
print(f"True Positive:{tp}") 
print(f"False Positive:{fp}") 
print(f"True Negative:{tn}") 
print(f"False Negative:{fn}") 
accuracy = 1.0 * (tp + tn) / (tp + tn + fp + fn) 
print(f"Accuracy:{accuracy:.4f}") 
print(f"Precision:{precision:.4f}") 
print(f"Recall:{recall:.4f}") 
print(f"F1 Score:{f1:.4f}")
```
